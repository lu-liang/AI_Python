```python
# please install nltk data before running this example. 

import nltk
nltk.download()
```

The nltk.download() will kick off the downloader of nltk as follows. Please download required data and then close the downloader to continue run this example. 



```python
sentence = '''At eight o'clock on Thursday morning Arthur didn't feel very good.'''
```

```python
tokens = nltk.word_tokenize(sentence)
tokens
```

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
entities = nltk.chunk.ne_chunk(tagged)
entities
```

```python
from nltk.corpus import treebank
t = treebank.parsed_sents('wsj_0001.mrg')[0]
t.draw()
```
