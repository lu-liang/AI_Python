```python
# define activation function
# 定义激活函数及其导数

import numpy as np
'''
tanh(x)=sinh(x)/cosh(x)
sinh(x)=(exp(x)-exp(-x))/2
cosh(x)=(exp(x)+exp(-x))/2
'''

def tanh(x):
    return np.tanh(x)

# tanh的导数
def tanh_deriv(x):
    return 1.0 - np.tanh(x)*np.tanh(x)

def logistic(x):
    return 1/(1 + np.exp(-x))

# logistic的导数
def logistic_derivative(x):
    return logistic(x)*(1-logistic(x))
```

```python
# generate random matrix 3 * 10
# define 3 data, each has 5 features. 
# 生成随机数据，显示
x = np.random.rand(3, 10)
print('x : {} \n'.format(x))

y_tanh = tanh(x)
print("tanh(x): {} \n ".format(y_tanh))

# show graph for data. 
import matplotlib
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, j, ax in zip(x, y_tanh, axes):
    ax.plot(i, j, '^', markersize=8)
    ax.set_xlabel("x")
    ax.set_ylabel("tanh(x)")
plt.show()

y_tanh_deriv = tanh_deriv(x)
print("tanh_deriv(x): {} \n ".format(y_tanh_deriv))
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, j, ax in zip(x, y_tanh_deriv, axes):
    ax.plot(i, j, 'v', markersize=8)
    ax.set_xlabel("x")
    ax.set_ylabel("tanh_deriv(x)")
plt.show()

y_logistic = logistic(x)
print("logistic(x): {} \n ".format(y_logistic))
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, j, ax in zip(x, y_logistic, axes):
    ax.plot(i, j, '+', markersize=8)
    ax.set_xlabel("x")
    ax.set_ylabel("y_logistic(x)")
plt.show()

y_logistic_derivative = logistic_derivative(x)
print("logistic_derivative(x): {} \n ".format(y_logistic_derivative))
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, j, ax in zip(x, y_logistic_derivative, axes):
    ax.plot(i, j, 'o', markersize=8)
    ax.set_xlabel("x")
    ax.set_ylabel("y_logistic_derivative(x)")
plt.show()
```

```python
# define neural network. 
# 定义神经网络
class NeuralNetwork:
    
    def __init__(self, layers, activation='tanh'):
        """
        :param layers: 
            A list containing the number of units in each layer. Should be at least two values
            for example, [3,3,1] means that there are 3 layers. The first layer is input layer which has 3 units.
            The second layer is hidden layer containing 3 units. The last layer is output layer.  
        :param activation: 
            The activation function to be used. Can be "logistic" or "tanh"
        """
        if activation == 'logistic':
            self.activation = logistic
            self.activation_deriv = logistic_derivative
        elif activation == 'tanh':
            self.activation = tanh
            self.activation_deriv = tanh_deriv
        
        # initial weight and bias
        # 初始化权重和偏向
        self.weights = []
        for i in range(1, len(layers) - 1):
            # + 1, represents the bias.
            self.weights.append((2*np.random.random((layers[i - 1] + 1, layers[i] + 1))-1)*0.25)
            self.weights.append((2*np.random.random((layers[i] + 1, layers[i + 1]))-1)*0.25)
            
    def __repr__(self):
        return ' activation: {}\n weights: {}\n'.format(self.activation, self.weights)
            
    def fit(self, X, y, learning_rate=0.2, epochs=10000):
        X = np.atleast_2d(X)
        temp = np.ones([X.shape[0], X.shape[1]+1])
        temp[:, 0:-1] = X
        X = temp
        y = np.array(y)

        for k in range(epochs):
            i = np.random.randint(X.shape[0])
            a = [X[i]]

            for l in range(len(self.weights)):
                a.append(self.activation(np.dot(a[l], self.weights[l])))
            error = y[i] - a[-1]
            deltas = [error * self.activation_deriv(a[-1])]

            for l in range(len(a) - 2, 0, -1):
                deltas.append(deltas[-1].dot(self.weights[l].T)*self.activation_deriv(a[l]))
            deltas.reverse()

            for i in range(len(self.weights)):
                layer = np.atleast_2d(a[i])
                delta = np.atleast_2d(deltas[i])
                self.weights[i] += learning_rate * layer.T.dot(delta)
                
    def predict(self, x):
        x = np.array(x)
        temp = np.ones(x.shape[0]+1)
        temp[0:-1] = x
        a = temp
        for l in range(0, len(self.weights)):
            a = self.activation(np.dot(a, self.weights[l]))
        return a
```

```python
# 应用上面定义的神经网络进行预测。

nn = NeuralNetwork([2,3,1], 'tanh')
print(nn)
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([1,0,0,1])
nn.fit(x,y,0.1,10000)
for i in [[0,0], [0,1], [1,0], [1,1]]:
    print(i, nn.predict(i))
    
print(nn)
```
